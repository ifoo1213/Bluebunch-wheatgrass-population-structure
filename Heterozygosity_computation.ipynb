```python
import allel
import pandas as pd
import numpy as np
```

```python
# load the VCF file
vcf_file = 'BBG_PI_snp_200_miss4mac3maf005.vcf'
callset = allel.read_vcf(vcf_file)

# Extract genotype data
genotypes = allel.GenotypeArray(callset['calldata/GT'])
```

```python
def compute_individual_heterozygosity(genotypes):
    # Initialize a list to store heterozygosity for each individual
    individual_heterozygosity = []

    # Iterate over each individual
    for i in range(genotypes.shape[1]):
        # Get the genotypes for the current individual
        individual_genotypes = genotypes[:, i]
        
        # Count the number of heterozygous sites
        het_count = np.sum(individual_genotypes.is_het())
        
        # Count the total number of genotyped sites (excluding missing data)
        total_genotyped = np.sum(individual_genotypes.is_called())
        
        # Compute heterozygosity for the individual
        heterozygosity = het_count / total_genotyped if total_genotyped > 0 else np.nan
        
        # Append the result to the list
        individual_heterozygosity.append(heterozygosity)
    
    return np.array(individual_heterozygosity)

# Compute heterozygosity for each individual
individual_heterozygosity = compute_individual_heterozygosity(genotypes)

# Display the results
print("Heterozygosity for each individual:")

# print(callset['samples'])
print(individual_heterozygosity)
```
